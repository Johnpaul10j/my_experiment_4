# The Ocean's Longest Diving Animals

How long can air-breathing animals hold their breath while diving? Which animals can stay submerged for the longest time?

The file `longest-diving-animals.csv` contains the recorded durations, in minutes, of the longest dives ever recorded in the scientific literature for different species of marine mammals. 

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('longest-diving-animals.csv')
df
```

### Project Ideas

- Create horizontal bar plots to show the longest diving animals, either for each category or within a category.

- Follow the Pro Tips.

- Convert dive durations from minutes to hours for easier understanding - 2.5 hours is clearer than 150 minutes.

- Add a reference bar to help readers build intuition about the durations. 

Side Note: The longest time a person has held their breath underwater is 24 minutes 37 seconds. This was done at the surface of a pool, not while freediving, so it might not directly comparable.


### Task

### Create a horizontal bar plot of the longest diving animals, converting dive durations from minutes to hours and adding a reference bar for the human breath-holding record.

## Data wrangling

### Subtask:
Convert the 'duration' column from minutes to hours.


### **Reasoning**:
## Create a new column 'duration_hours' by converting the 'duration' column from minutes to hours.



```python
df['duration_hours'] = df['duration'] / 60
df
```

## Data visualization

### Subtask:
Create a horizontal bar plot of the longest diving animals, highlighting the top animals and including a reference line for the human breath-holding record.


## **Reasoning**:
Sort the dataframe, select the top animals, and create a horizontal bar plot with a reference line for the human breath-holding record, including appropriate labels and title.



```python
df_sorted = df.sort_values('duration_hours', ascending=False)
top_n = 15
df_top = df_sorted.head(top_n)

human_record_minutes = 24 * 60 + 37
human_record_hours = human_record_minutes / 3600

plt.figure(figsize=(10, 8))
plt.barh(df_top['animal'], df_top['duration_hours'], color='skyblue')
plt.axvline(x=human_record_hours, color='red', linestyle='--', label=f'Human Record ({human_record_hours:.2f} hours)')

plt.xlabel('Dive Duration (hours)')
plt.ylabel('Animal')
plt.title('Top Longest Diving Animals and Human Breath-Holding Record')
plt.legend()
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

## Summary:

### Data Analysis Key Findings

* A new column, 'duration_hours', was successfully created by converting the original 'duration' column from minutes to hours.
* The top 15 longest diving animals based on their dive duration in hours were identified and visualized.
* The human breath-holding record was calculated to be approximately 0.41 hours and included as a reference line on the plot.
* The plot visually demonstrates how the dive durations of the top diving animals compare significantly to the human breath-holding record.

### Insights or Next Steps

* The large difference in dive durations between top diving animals and the human record highlights the extreme physiological adaptations for diving in these animals.
* Investigating other physiological factors (e.g., oxygen storage, metabolic rate) in these animals could provide deeper insights into the mechanisms behind their extended dive capabilities.
